```python
import pandas as pd
import numpy as np

train_datasets= pd.read_csv('xtest.csv')
half_count = len(train_datasets) / 2
train_datasets = train_datasets.dropna(thresh=half_count,axis=1) # Drop any column with more than 50% missing values
```

```python
train_datasets.head(5)
```

```python
train_datasets.shape
```

```python
# train_datasets = train_datasets.dropna()
train_datasets.fillna(train_datasets.mean(), inplace=True)
null_counts = train_datasets.isnull().sum()
print("Number of null values in each column:\n{}".format(null_counts))
```

```python
train_datasets.shape
```

```python
print("Data types and their frequency\n{}".format(train_datasets.dtypes.value_counts()))
```

```python
train_datasets_df = train_datasets.select_dtypes(include=['object'])
print(train_datasets_df.iloc[0])
```

```python
cols = ['Month', 'VisitorType']
for name in cols:
    print(name,':')
    print(train_datasets_df[name].value_counts(),'\n')
```

```python
mapping_dict = {'Month':
                {'Jan':1,'Feb':2,'Mar':3,'April':4,'May':5,'June':6,'Jul':7,'Aug':8,'Sep':9,'Oct':10,'Nov':11,'Dec':12}
               }
train_datasets= train_datasets.replace(mapping_dict)
```

```python
train_datasets.head(5)
```

```python
nominal_columns = ["VisitorType"]
dummy_df = pd.get_dummies(train_datasets[nominal_columns])
train_datasets = pd.concat([train_datasets, dummy_df], axis=1)
train_datasets = train_datasets.drop(nominal_columns, axis=1)
```

```python
train_datasets.head(5)
```

```python
train_datasets['Weekend'] = train_datasets.Weekend.astype(int)
```
